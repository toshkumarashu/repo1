# Movie Recommendation System

Recommendation System, also labelled as recommendation systems, are statistical algorithms that recommend products to users based on similarities between the buying trends of various user or similarities between the products.

Collaborative Filtering- The process used to calculate similarities between the buying trends of various users or similarities between products is called collaborative filtering.

User based collaborative filtering= If two user X and Y, like products A and B and there is another user X who likes product A, then the procuct B will also be recommended to user Z.

Item-based collaborative filtering- In this, products are recommended based on similarties between themselves. For instance if a user like product A and product A has properties X and Y, another product B with properties X and Y will also be recommended to the user.


# Item Based Collaborative Filtering

In item based collaborative filterig, products are recommended based on common characteristics.

The first step is to create a dataframe where each movie is represented by a column and rows contain user rating for movies.

```python
user_movie_rating_matrix=merged_movie_df.pivot_table(index="userId",columns="title",values="rating")
user_movie_rating_matrix
```

```python
user_movie_rating_matrix.shape
```


The Dataset contains 610 unique users and 9719 unique movies.

Now we will find the movie recommendation based on a single movie and then based on multiple movies.

Finding recommendations based on a single movie- Suppose we want to find the recommendation based on the movie Pulp Fiction.

First we will filter the column that contains the user ratings for the movie.

```python
pulp_fiction_ratings=user_movie_rating_matrix["Pulp Fiction (1994)"]
```

Next, we will find the correlation between the user ratings of all the movies and the user ratings for the movie pulp fiction.

```python
pulp_fiction_correlations=pd.DataFrame(user_movie_rating_matrix.corrwith(pulp_fiction_ratings),columns=["pf_corr"])
```

```python
pulp_fiction_correlations.sort_values("pf_corr",ascending=False).head(5)
```


Correlation itself is not giving meaningful results. One solution to this probleem can be that in addition to the correlation between the movies, we also use rating counts, for the correlated movie as criteria for finding the best recommendation

```python
pulp_fiction_correlations=pulp_fiction_correlations.join(movie_rating_mean_count["rating_count"])
pulp_fiction_correlations.head()
```


The pf_corr column contains some NaN vlues. This is because there can be movies that are rated by users who did not rate Pulp Fiction (1994). In such cases, correlation will be null

We will remove all movies with null correlation with Pulp Fiction(1994).


```python
pulp_fiction_correlations.dropna(inplace=True)
```

```python
pulp_fiction_correlations.sort_values("pf_corr",ascending=False).head()
```


A better way is to find the movies with the rating count of atleast 50 and having the highest correlation with Pulp Fiction(1994).

```python
pulp_fiction_correlations_50=pulp_fiction_correlations[pulp_fiction_correlations['rating_count']>50]
pulp_fiction_correlations_50.sort_values("pf_corr",ascending=False).head()
```

Finding the recommendation based on multiple movies- In this we will see how to recommend movies to a user based on his ratings of multiple movies. The first step is to create a dataframe, which contains a correlation between all the movies in our dataset in the form of matrix.

```python
all_movie_correlations=user_movie_rating_matrix.corr(method="pearson",min_periods=50)
```

```python
all_movie_correlations.head()
```


Now suppose a user logs into the website.The user has already watched three movies and given ratings to those movies.


```python
movie_data=[['Forrest Gump (1994)',4.0],['Fight Club (1999)',3.5],['Interstellar (2014)',4.0]]
test_movies=pd.DataFrame(movie_data,columns=['Movie_Name','Movie_Rating'])
test_movies.head()
```

```python
print(test_movies['Movie_Name'][0])
print(test_movies['Movie_Rating'][0])
```


From all_movie_correlation dataframe, let's obtain correlaion values for the movies related to Forrest Gump(1994)

```python
all_movie_correlations['Forrest Gump (1994)'].dropna()
```


Next, we will iterate through the three movies in the test_movie dataframe, find the correlated movies, and then multiply the correlation of all the correlated movies with the ratings of the input movie.

The correlated movies, along with the weighted correlation are appended to an empyt series named recommended movies.

```python
recommended_movies=pd.Series()

for i in range(0,2):
  movie=all_movie_correlations[test_movies['Movie_Name'][i]].dropna()
  movie=movie.map(lambda movie_corr:movie_corr*test_movies["Movie_Rating"][i])
  #recommended_movies=recommended_movies.append(movie)
  recommended_movies = pd.concat([recommended_movies, movie])
```

```python
recommended_movies
```


TO get a final recommendation, you can sort the movies in the ascending order of the weighted correlation

```python
recommended_movies.sort_values(inplace=True,ascending=False)
recommended_movies.head(10)
```
